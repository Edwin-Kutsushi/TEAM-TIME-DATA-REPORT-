## 1.1 Identify the number of Devices per disability Category

To achieve this, we classified the devices into 6 categories as follows:
*  Lame/Missing_Hand/Missing_Foot
*  Paralysis
*  Dumb
*  Deaf
*  Mental
*  Blind

This was inspired by the classification of disabilities in the 'Percentage_Distribution_of_Population_by_type_of_Disability_County_Estimates%2C2005_6' dataset which had the same classification as above.

```python
# Creating a new column that will contain the classification of the devices per disability. 
Assistive_Devices['Disability_type'] = Assistive_Devices['disability']

# Classifying the disabilities into: 'Lame/Missing_Hand/Missing_Foot', 'Paralysis', 'Dumb', 'Other', 'Deaf', 'Mental', 'Blind'

conditions = [
    (Assistive_Devices['Disability_type'] == 'Physical'),
    (Assistive_Devices['Disability_type'] == 'Cerebral Palsy'),
    (Assistive_Devices['Disability_type'] == 'Physical/Visual/Mental'),
    (Assistive_Devices['Disability_type'] == 'Physical/Speech'),
    (Assistive_Devices['Disability_type'] == 'Multiple'),
    (Assistive_Devices['Disability_type'] == 'Spinabifida'),
    (Assistive_Devices['Disability_type'] == 'Physical/Hearing'),
    (Assistive_Devices['Disability_type'] == 'Hdrocephalus'),
    (Assistive_Devices['Disability_type'] == 'Physical/Cerebral Palsy'),
    (Assistive_Devices['Disability_type'] == 'Mental'),
    (Assistive_Devices['Disability_type'] == 'Physical/Mental'),
    (Assistive_Devices['Disability_type'] == 'Physical,Speech'),
    (Assistive_Devices['Disability_type'] == 'Intellectual'),
    (Assistive_Devices['Disability_type'] == 'Physical/Visual'),
    (Assistive_Devices['Disability_type'] == 'Speech'),
    (Assistive_Devices['Disability_type'] == 'Spina Bifida & Hydrocephalus')
    ]

values = [
    'Lame/Missing_Hand/Missing_Foot',
    'Paralysis',
    'Lame/Missing_Hand/Missing_Foot',
    'Dumb',
    'Other',
    'Paralysis',
    'Deaf',
    'Paralysis',
    'Paralysis',
    'Mental',
    'Mental',
    'Dumb',
    'Mental',
    'Blind',
    'Dumb',
    'Paralysis'
]

Assistive_Devices['Disability_type'] = np.select(conditions, values)
```

```python
# Using group by and count to identify the number of devices per disability category
no_of_devices_per_category = pd.DataFrame(Assistive_Devices['Disability_type'].groupby(Assistive_Devices['Disability_type']).agg(['count']))

# Re-indexing
no_of_devices_per_category['Type_of_disability'] = no_of_devices_per_category.index
no_of_devices_per_category.reset_index(drop=True, inplace=True)

# Renaming columns
no_of_devices_per_category.columns = ['Total_number_of_disability_devices','Type_of_disability']

# Re-arranging the columns
no_of_devices_per_category = no_of_devices_per_category[['Type_of_disability','Total_number_of_disability_devices']]

no_of_devices_per_category.sort_values(by="Total_number_of_disability_devices", ascending = 0)
```

## 2. Distribution of devices per county and checking the percentage of the counties that do not have the specific devices.

**2.1 Counties that did not receive any disability devices.**

```python
# Using a left join to get the counties that are only in the 'Percentage_Disability_per_County' dataset.
counties_without_devices = pd.merge(Percentage_Disability_per_County, Assistive_Devices, on='county', how='left')
counties_without_devices = counties_without_devices[counties_without_devices['disability'].isnull()]
counties_without_devices = pd.DataFrame(counties_without_devices['county'])
counties_without_devices.reset_index(drop=True, inplace=True)
counties_without_devices
```

**2.3 Obtain the percentage of the counties with Lame/Missing_Hand/Missing_Foot disabilities and without 'Lame/Missing_Hand/Missing_Foot' devices .**

```python
# 1. Getting the total number of people with Lame/Missing_Hand/Missing_Foot disabilities per county from the Percentage_Disability_per_County dataset
Percentage_Disability_per_County['Lame_Missing_Hand_Missing_Foot'] = Percentage_Disability_per_County['missing_hand'] + Percentage_Disability_per_County['missing_foot'] + Percentage_Disability_per_County['lame']

# 2. Extracting the counties that have people living with Lame/Missing_Hand/Missing_Foot disabilities from the Percentage_Disability_per_County dataset
counties_with_leg_and_foot_disabilities = Percentage_Disability_per_County[Percentage_Disability_per_County['Lame_Missing_Hand_Missing_Foot'] != 0]

# 3. Extract the counties with Lame/Missing_Hand/Missing_Foot disability devices
counties_with_leg_and_foot_devices = Assistive_Devices[Assistive_Devices['Disability_type'] == 'Lame/Missing_Hand/Missing_Foot']
counties_with_leg_and_foot_devices = pd.DataFrame(counties_with_leg_and_foot_devices['county'].unique())
counties_with_leg_and_foot_devices

# 4. Getting the total number of counties that have people living with Lame/Missing_Hand/Missing_Foot disabilities and do not have the corresponding devices
counties_without_leg_and_foot_devices = counties_with_leg_and_foot_disabilities['county'].count() - counties_with_leg_and_foot_devices.count()
counties_without_leg_and_foot_devices

# 5. Getting the percentage of counties that have people living with Lame/Missing_Hand/Missing_Foot disabilities and do not have the corresponding devices
percentage_counties_without_leg_and_foot_devices = (counties_without_leg_and_foot_devices/counties_with_leg_and_foot_disabilities['county'].count()) * 100
print("Out of the " + str(counties_with_leg_and_foot_disabilities['county'].count()) +" counties that had Lame/Missing_Hand/Missing_Foot disabilities, "+ str(round(percentage_counties_without_leg_and_foot_devices[0],0)) + "% of them did not have the corresponding devices")
```
